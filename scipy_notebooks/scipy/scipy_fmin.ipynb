```python
import numpy as np
```

```python
def f(x):
    return (np.log(x)-2.)**2+1.
```

```python
x=np.linspace(4,10,100)
```

```python
y=f(x)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
plt.plot(x,y)
```

```python
from scipy import optimize as opt
xsol=opt.fmin(func=f,x0=4.)

print(xsol)
```

```python
plt.plot(x,y)
plt.axhline(f(xsol[0]),c='g')
plt.axvline(xsol[0],c='g')
plt.scatter(xsol,f(xsol),c='r')
```
