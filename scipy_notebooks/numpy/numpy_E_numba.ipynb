```python
import numpy as np
def bubblesort(X):
    N = X.shape[0]
    for end in range(N, 1, -1):
        for i in range(end - 1):
            if X[i] > X[i + 1]:
                X[i], X[i + 1] = X[i + 1], X[i]

from numba import jit
bubblesort_jit = jit(nopython=True)(bubblesort)
```

```python
x=np.random.rand(99)
print(x)
bubblesort(x)
print(x)
```

```python
%%timeit
bubblesort(np.random.rand(99))
```

```python
%%timeit
bubblesort_jit(np.random.rand(99))
```
