```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,100)
y=np.linspace(0,10,100)
xmesh,ymesh=np.meshgrid(x,y)
z=np.sin(xmesh)*np.cos(ymesh)
z
```

```python
plt.contourf(z)
plt.colorbar()
plt.contour(z,levels=[0],colors=['k'],linestyles=[':'])
```
